# Normalise the data

```python
columns_not_normalised=['Class']
columns_to_normalise=[x for x in df.columns if x not in columns_not_normalised]
for i in columns_to_normalise:
    df[i] = ( df[i] - df[i].mean() ) / df[i].std()
df.head()
```

# Divide and store data into test and train sets.

```python
x_columns=[x for x in df.columns if x != 'Class']
y_column='Class'

X=df[x_columns]
Y=df[y_column]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=r_state)

train_data=pd.concat([X_train,Y_train],axis=1)
test_data=pd.concat([X_test,Y_test],axis=1)

train=train_data.to_csv('train.csv')
test=test_data.to_csv('test.csv')


```

# Statistical analysis

Pearson's Correlation

=> correlation close to 1 or -1 is good.

=> higher prob_value determines how accurate correlation is.

=> Nothing interesting is achieved as class is not a numeric data. It is a categorical data that too biased.

```python
dependent_feature='Class'
independent_features=[x for x in df.columns if x != dependent_feature]
for i in independent_features:
    pearsons_coef,prob_value=stats.pearsonr(df[i],df[dependent_feature])
    print(i,": pearsons correlation coefficient = ",pearsons_coef," probability value = ",prob_value)
print("__________________________________________________________________________________________________________________")
print()
```

Analysis of variance(ANOVA)

=>Large f implies stronger correlation

=>Smaller the p more accurate is the f value. 

=>Columns v1 to v12, v14, v16 to v21 are highly correlated to column Class 

```python
df1=df
df1['Class']=df1['Class'].apply(str)
dependent_feature='Class'
independent_features=[x for x in df.columns if x != dependent_feature]
for i in independent_features:
    df_for_anova=df1[[i,dependent_feature]]
    grouped_for_anova=df_for_anova.groupby(dependent_feature)
    f,p=stats.f_oneway(grouped_for_anova.get_group('1')[i],grouped_for_anova.get_group('0')[i])
    print(i,": f-test = ",f," probability value = ",p)
```

```python
useful_features=[x for x in df.columns if x not in ['Time','V13','V15','V22','V23','V24','V25','V26','V27','V28','Amount']]
print(useful_features)
```

# Model 

train validation split

```python
df=pd.read_csv('train.csv')
X=df[useful_features]
Y=df['Class']

X_train, X_dev, Y_train, Y_dev = train_test_split(X, Y, test_size=0.18, random_state=r_state)
```

Outlier fraction

```python

state = np.random.RandomState(24)

valid=Y_train[Y_train==0]
fraud=Y_train[Y_train==1]
outlier_fraction=len(fraud)/float(len(valid))
```

Model and it's fit

```python
model=IsolationForest(n_estimators=100, max_samples=len(X_train), contamination=outlier_fraction,random_state=state, verbose=0)
model.fit(X_train)
scores_prediction = model.decision_function(X_train)
y_train_pred = model.predict(X_train)
```

Prediction formatting and training evalution

```python
y_train_pred[y_train_pred == 1] = 0
y_train_pred[y_train_pred == -1] = 1
n_errors = (y_train_pred != Y_train).sum()


print("\n\nTraining")
print("__________________________________________________________________________________________\n")
print("Classification Report :")
print(classification_report(Y_train,y_train_pred))
print("__________________________________________________________________________________________\n")
print("Confusion matrix")
print(confusion_matrix(Y_train,y_train_pred))
print("__________________________________________________________________________________________\n")
print("error: ",n_errors) 
print("Accuracy Score :",Fore.RED+str(accuracy_score(Y_train,y_train_pred)))
```

Prediction, formatting and validation

```python
y_dev_pred = model.predict(X_dev)
y_dev_pred[y_dev_pred == 1] = 0
y_dev_pred[y_dev_pred == -1] = 1
n_errors = (y_dev_pred != Y_dev).sum()


print("\n\nValidation")
print("_________________________________________________________________________________\n")
print("Classification Report :")
print(classification_report(Y_dev,y_dev_pred))
print("_________________________________________________________________________________\n")
print("Confusion matrix")
print(confusion_matrix(Y_dev,y_dev_pred))
print("_________________________________________________________________________________\n")
print("error: ",n_errors) 
print("Accuracy Score :",Fore.RED+str(accuracy_score(Y_dev,y_dev_pred)))
```

# Testing

Prediction and evaluation for testing dataset

```python
df1=pd.read_csv('test.csv')
X_test=df1[useful_features]
Y_test=df1['Class']

y_test_pred = model.predict(X_test)
y_test_pred[y_test_pred == 1] = 0
y_test_pred[y_test_pred == -1] = 1
n_errors = (y_test_pred != Y_test).sum()


print("\n")
print("Testing evalutaion")
print("_____________________________________________________________________________________________\n")
print("Confusion matrix")
print(confusion_matrix(Y_test,y_test_pred))
print("_____________________________________________________________________________________________\n")
print("Classification Report :")
print(classification_report(Y_test,y_test_pred))
print("_____________________________________________________________________________________________\n")
print("error: ",n_errors) 
print("Accuracy Score: ",Fore.RED+str(accuracy_score(Y_test,y_test_pred)))
```
